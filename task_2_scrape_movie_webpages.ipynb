## Function: make_wiki_url(relative_path)


```python
def make_wiki_url(relative_path, base_path = 'https://en.wikipedia.org/'):
    url= base_path + relative_path
    return url
```

```python
movie_wiki_link_list = [make_wiki_url(movie['href']) for movie in title_href_list]
movie_wiki_link_list
```

## Function: make_table_rows(soup)


```python
def make_table_rows(soup):
    table_rows = soup.select('table.infobox.vevent tr')
    return table_rows
```

## Function: get_info_boxes(href_list)

```python
# create a function that takes in a list of relative wiki path url's.
# create wiki url paths with make_wiki_url(relative_path, base_path = "https://en.wikipedia.org/")
# create soup objecs using make_soup(url) for each link in make_wiki_link_list
# create a list of info_boxes using make_table_rows(soup) for each soup in soups (list).
# return back the list of info_boxes
def get_info_boxes(href_list):
    
    movie_wiki_link_list=[make_wiki_url(movie['href']) for movie in href_list]
    
    soups = [make_soup(url) for url in movie_wiki_link_list]
    
    info_boxes = [make_table_rows(soup) for soup in soups]
    
    return info_boxes
    
```

```python
# create all the info boxes
info_boxes = get_info_boxes(title_href_list)
```

```python
# [clean]
# get rid of subscripts ([1],[2],etc.)

def remove_tags(content):
    # [clean] removing subscripts:
    if content.find_all(['sup','span']):
        for tag in content.find_all('sup'):
            tag.decompose()
                        
    content = content.get_text(" ", strip=True).replace('\xa0', ' ')
    
    return content
```

## Function: get_br_list(td), get_content_value(content)

```python
def get_br_list(td):

    if len(td.find_all('br')) > 1: 
        content_value = ([str(td.find_all('br')[x].previousSibling) for x in range(len(td.find_all('br')))])
        content_value.append(str(td.find_all('br')[-1].next_sibling))
    elif len(td.find_all('br')) == 1 :
        a = []
        [(a.append(str(td.find_all('br')[x].previousSibling)), a.append(str(td.find_all('br')[x].nextSibling))) for x in range(len(td.find_all('br')))]
        content_value=a
    
    return [x.lower() for x in content_value]
```

```python
def get_content_value(content):
    # find the <td> element in the bs element
    content_value = content.find('td')

    # if the <td> element contains list elements <li>: then put the text from those elements into a list named content_value
    # remove the subscripts from any list elements using remove_tags(content)
    if content_value.find('li'):
        li_list = content_value.find_all('li')
                    
        # [clean] subscripts out of li items
        content_value = [remove_tags(x).lower() for x in li_list]
    
    # if there are <br> elements and have links <a href=>: store them in a list named content_value.
    # else: store the value in a variable named content_value
    elif content_value.find('br'):
        # ['clean'] get long strings into a list. 
        # These items were not storring as lists such as in the 'Starting' key
        
        if content_value.find('a'):     
            a_list = content_value.find_all('a')
            content_value= [remove_tags(x).lower() for x in a_list]
        else:
            content_value= get_br_list(content_value)

#             content_value = remove_tags(content_value)
    # Else: get the text from the table td.                                
    else:
        content_value=remove_tags(content_value).lower()
        
    return content_value
```

```python
# Create a dictionary that contains the keys and values of movie information
def create_movie_dictionary(row_data):
    movie_info={}
    for index, row in enumerate(row_data):
        if index==0:
            movie_info['title'] = row.find("th").get_text().lower()
        elif index > 1:
            try:
                content_key = row.find('th').get_text(" ", strip=True)

                content_value = get_content_value(row)

                movie_info.update({content_key:content_value})

            except Exception as e:
                continue
        
    return movie_info
```

## Function: write to json(obj, file_name), load_data(file_name)

```python
# Save list of Dictionaries as a json file.
import json

def write_to_json(obj, file_name):
    with open(file_name, 'w') as fout:
        json.dump(obj, fout, ensure_ascii=False, indent=2)

def load_data(file_name):
    with open(file_name, encoding='utf-8') as f:
        return json.load(f)
```

```python
# write df to a .json file.
write_to_json(df, 'movie_dictionaries.json')
```

```python
# load 'movie_dictionaries.json' and store as a variable named df. 
# If 'movie_dictionaries.json' does not exist in the directory this function will fail.
df = load_data('movie_dictionaries.json')
```
