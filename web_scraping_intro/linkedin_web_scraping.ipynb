```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.linkedin.com/jobs/search?keywords=Data%20Science&location=San%20Antonio%2C%20Texas%20Metropolitan%20Area&geoId=90000724&trk=public_jobs_jobs-search-bar_search-submit&redirect=false&position=1&pageNum=0'
```

```python
url
```

```python
r = requests.get(url)
```

```python
webpage=bs(r.content)
```

```python
print(webpage.prettify())
```

```python
card_contents = webpage.select('ul.jobs-search__results-list')
```

```python
print(card_contents[0].prettify())
```

```python
len(card_contents)
```

```python
title_list = card_contents[0].select('span.screen-reader-text')
title_list
```

```python
snippet_list = card_contents[0].select('p.job-result-card__snippet')
snippet_list
```

```python
job_dictionary={'snippit_list':[], 'title_list':[]}
```

```python
job_dictionary
```

```python
[job_dictionary['snippit_list'].append(x.get_text()) for x in snippet_list]
```

```python
job_dictionary
```

```python
[job_dictionary['title_list'].append(x.get_text()) for x in title_list]
```
